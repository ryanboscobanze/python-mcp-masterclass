# sqlite3 Basics

Prerequisite for the MCP lifespan/sqlite module, which manages a real SQLite database via the server lifespan hook.

The file walks through 9 things you need to read that code:

1. **Connect** — `sqlite3.connect(":memory:")` for in-memory, or a file path for persistent
2. **Create and seed** — `executescript()` runs multiple SQL statements in one call
3. **fetchall()** — returns results as a list of tuples, accessed by index
4. **sqlite3.Row** — set `row_factory` to access columns by name instead of index, and convert rows to `dict`
5. **fetchone()** — read a single row
6. **Parameterised queries** — use `?` placeholders, never string formatting, to prevent SQL injection
7. **Error handling** — `sqlite3.Error` is the base class; one `except` block catches all SQL errors
8. **List tables** — query `sqlite_master` to inspect the schema
9. **Close** — `conn.close()` in the lifespan module runs in a `finally` block so it always runs
